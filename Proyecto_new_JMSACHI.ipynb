**FUNDAMENTACIÓN DEL PROYECTO**

Nuevo replanteo del problema de la predicción de condiciones ideales para un cambio de currícula digital desde regresión logística 

Este proyecto plantea la posibilidad de realizar un modelo, que alimentado por ciertos datos parametrizados, sea capaz de indicar la necesidad de actualización de curricular acrecentando contenidos digitales en los diversos niveles educativos.
        La pandemia del COVID 19 convirtió al mundo en un gran laboratorio de aceleramiento de la “transformación digital” en todos los medios y en particular el educativo, mostrando las potencialidades pero también las limitaciones o brechas, que son consecuencias de las inequidades sociales y culturales arrastradas del pasado que significan imposibilidad al acceso tecnológico. Esta brecha tecnológica que se inicia en la baja probabilidad de poder acceder a equipamiento (computadora, celular etc.) y flujo de datos dado por el acceso nulo a internet. Esto produce una atrofia en el desarrollo de las denominadas “competencias digitales”. Los gobiernos del mundo se están viendo forzados a revertir esta situación y toda herramienta es útil si ayuda a esto en cualquier sentido. De aquí la necesidad de un modelo que determine la necesidad de cambios de curriculas con el propósito de su digitalización.
        Una competencia digital se define como el uso crítico y seguro, mediante las tecnologías de la sociedad de la información para el trabajo, el ocio y la comunicación. Supone un conjunto de conocimientos, habilidades, actitudes y estrategias que se requieren para el uso de los medios digitales y de las tecnologías de información y comunicación. La competencia digital se apoya en las habilidades del uso de ordenadores para “recuperar, evaluar, almacenar, producir, presentar e intercambiar información y para comunicar y participar en redes de colaboración a través de Internet” (European Parliament and the Council, 2006). Para parametrizar lo que se entienda como como competencia digital deberá incluir datos de la accesibilidad a equipos y flujo de datos.  
           Los docentes han respondido a la desafiante situación del COVID-19, pudiendo identificar prácticas digitales valiosas deseables en el futuro en todas las esferas de la vida. Escuelas y administración educativa pueden haber proporcionado un apoyo valioso. Pero además expresaron preocupaciones con respecto a sus alumnos, teniendo la falta de acceso a alguna de ellas. 
          Varios estudios no indican mucha inercia y resistencia hacia esta transformación. Una transformación tan dramática no da lugar a eso. Se plantearon algunos problemas relacionados con la actitud en tanto a niños como a sus padres. En definitiva y  en consonancia con la investigación previa sobre transformación digital, parece que los tecnología, prácticas, habilidades y competencias, actitudes y aspectos culturales pueden estar actuando como barreras a la transformación digital. Donde se entremezclan cuestiones culturales, valores, actitudes, mentalidades e identidades. La única actitud aceptable,que debe adoptar el estado es una postura proactiva y motivadora hacia el diseño y la tecnología digital; pues así se se apoya la innovación y diseño de mejores herramientas para satisfacer las necesidades de la educación básica digitalizada. Esto es algo que debería aparecer si empiezan a prevalecer los arreglos de educación a distancia, por ejemplo. Todo esto implica necesariamente fuertes cambios en las curriculas y obliga fomentar y aportar al cierre de las brechas mencionadas.
           Históricamente, para la parametrización de la curricula se recurrió a su monitoreo mediante evaluaciones que pusieran en evidencia el grado de logros de los objetivos y propósitos planteados en ellas. Es,indudablemente, un método eficaz que debe ser tomado en cuenta.

````python
import statsmodels.formula.api as smf
````

````python
b= db[["Uso_adecuado_tecnologia", "Comunidades_TI_especializadas", "Proyectos_TI"]]
lm = smf.ols(formula="Resuelvo_prob_tecnicos~b", data = db ).fit()
````

````python
lm.params
````

````python
lm_pred = lm.predict(pd.DataFrame(db["Uso_adecuado_tecnologia"]))
lm_pred
````

````python
%matplotlib inline
db.plot(kind = "scatter", x = "Uso_adecuado_tecnologia", y = "Resuelvo_prob_tecnicos")
plt.plot(pd.DataFrame(db["Uso_adecuado_tecnologia"]), lm_pred, c="red", linewidth = 2)
````
